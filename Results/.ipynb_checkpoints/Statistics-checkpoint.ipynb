```python
import pandas as pd
import numpy as np

df = pd.read_csv('agree100.csv', index_col = ['_unit_id'])
df
```

```python
def count(x):
    
    html = x['html']
    check = x['sentena_sem_relao']
    
    if check:
    
        dictStatus['sem relação'] += 1
        
    elif html.find('class=\'per\'') >= 0:
        
        dictStatus['org x per'] += 1
    
    elif html.find('class=\'plc\'') >= 0:
        
        dictStatus['org x plc'] += 1
    
    else: 
        
        dictStatus['org x org'] += 1
        
        
dictStatus = {'org x org': 0, 'org x per': 0, 'org x plc': 0, 'sem relação': 0}
df.apply(count,axis = 1)

print(dictStatus)
```
