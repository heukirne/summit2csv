```python
import pandas as pd
import numpy as np

df = pd.read_csv('f1027749.csv', index_col = ['_unit_id'])
df
```

```python
def annotations(x):
    #print(x.name)
    
    if not x.name in dictAnnotations:
        
        #print(x['_golden'])
        dictAnnotations[x.name] = []
    
    dictAnnotations[x.name].append(x['_worker_id'])
    
    
def biggest(x):
    
    biggest = 0
    
    for key in x:
        
        if x[key] >= biggest:
            
            biggest = x[key]
            theBiggest = key
    
    return theBiggest 
    
    
def translations(x):
    
    kappa = [0, 0]
    
    for y in x:
        
        if y != '-1':
            
            kappa[0] += 1
        
        else:
            
            kappa[1] += 1
    
    return kappa
    

dictAnnotations = {}
df.apply(annotations,axis = 1)

#print(dictAnnotations)

dictFrequency = {}

for key in dictAnnotations:
    
    aux = tuple(dictAnnotations[key])
    
    if not aux in dictFrequency:
        
        dictFrequency[aux] = 1
        
    else:
        
        dictFrequency[aux] += 1
        
#print(dictFrequency)

theBiggest = str(biggest(dictFrequency)).replace("(", "").replace(")", "").split(",")
one = theBiggest[0]
two = theBiggest[1]
three = theBiggest[2]

#print(theBiggest)

idSentences = []
notIdSentences = []

for key in dictAnnotations:
    
    if str(dictAnnotations[key]).find(one) < 0 or str(dictAnnotations[key]).find(two) < 0 or str(dictAnnotations[key]).find(three) < 0:
        
        notIdSentences.append(key)
    
    else:
        
        idSentences.append(key)
        
#print(idSentences)

dff = df.drop(notIdSentences)
dff = dff.reset_index()
#dff

matrix = []

for x in idSentences:
    
    aux = dff.loc[dff['_unit_id'] == x]
    aux['id_rel']
    matrix.append(aux['id_rel'].tolist())
    
#print(matrix)

kappa = []
    
for x in matrix:

    kappa.append(translations(x))
    
print(kappa)

from statsmodels.stats.inter_rater import fleiss_kappa

fleiss_kappa(kappa)
```
