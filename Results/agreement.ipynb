```python
import pandas as pd
import numpy as np

array100 = []
array66 = []

df = pd.read_csv('a1027749.csv',index_col='_unit_id')

def agreement(x):
    lista = x['id_rel'].split(' ')
    
    if lista[0] == lista[1]:
        if lista [0] == lista[2]:
            return 100
        return 66
    
    if lista[1] == lista[2] or lista[0] == lista[2]:
        return 66
    return 33
    

df['agree'] = df.apply(agreement, axis=1)

df[['id_rel','agree']]
```

```python
df100 = df[df['agree'] == 100]
df100.to_csv('agree100.csv')

print(df100.shape)

df66 = df[df['agree'] == 66]
df66.to_csv('agree66.csv')

print(df66.shape)

df33 = df[df['agree'] == 33]
df33.to_csv('agree33.csv')

print(df33.shape)
```
