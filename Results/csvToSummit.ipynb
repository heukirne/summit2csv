```python
import pandas as pd
import numpy as np

df = pd.read_csv('agree100.csv')
df.sample(2)
```

```python
def id1_min(x):
    return x['id_ent1'].split('-')[0]
    
def id1_max(x):
    try:
        return x['id_ent1'].split('-')[1]
    except:
        return x['id_ent1'].split('-')[0]
    
def id2_min(x):
    return x['id_ent2'].split('-')[0]

def id2_max(x):
    try:
        return x['id_ent2'].split('-')[1]
    except:
        return x['id_ent2'].split('-')[0]
    
df['id_ent1_min'] = df.apply(id1_min, axis=1)
df['id_ent1_max'] = df.apply(id1_max, axis=1)
df['id_ent2_min'] = df.apply(id2_min, axis=1)
df['id_ent2_max'] = df.apply(id2_max, axis=1)
```

```python
open("Summ-it++v3_relation_gold.txt", 'w').close()
with open("/home/pucrspln/summit2csv/Summ-it++ Corpus and csv conversions/Summ-it++v3.txt", "r", encoding = "utf8") as summit, open("Summ-it++v3_relation_gold.txt", "a") as saida:

    def id_rel(x):
        return x['id_rel'].split(' ')[0].replace(',','|')
    
    for line in summit:
        
        if line.find("#begin") >= 0:
            columns = line.split('/')
            idText = columns[7]
            
        columns = line.split('\t')
        
        if len(columns) >= 6:
            idSent = columns[0]
            idToken = columns[1]
            
            relations = df.loc[(df['id_txt'] == idText) 
                               & (df['id_sent'] == idSent) 
                               & ((df['id_ent1_min'] == idToken) | (df['id_ent1_max'] == idToken) | (df['id_ent2_min'] == idToken) | (df['id_ent2_max'] == idToken)) 
                               & (df['_golden'] == False) 
                               & (df['sentena_sem_relao'] == False) 
                               & (df['_unit_state'] == 'finalized')
                              ]
            
            if not relations.empty:
                idRel = relations.apply(id_rel,axis = 1)
                columns[8] = '(' + ')('.join(idRel.values) + ')'
                line = '\t'.join(columns)
                
                print(line)
                
        #saida.write(line)
```
