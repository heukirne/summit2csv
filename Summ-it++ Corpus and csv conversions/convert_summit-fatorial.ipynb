```python
def montaTexto(idText, idSent, k1, k2, v1, v2, palavras): 
    idData = ""
    strWord = ""
    strLine = ""
    check = False
    idDataK1 = str(k1)
    idDataK2 = str(k2)
    for idx in palavras:
        
        # monta IDs compostos
        if palavras[idx][1].find("(") >= 0 or len(palavras[idx][1]) == 1:
            idData += str(idx)
        elif palavras[idx][1].find(")") >= 0:
            idData += "-" + str(idx)
            # salva IDs compostos das entidades marcadas
            if idData.find(str(k1)+'-') == 0:
                idDataK1 = idData
            if idData.find(str(k2)+'-') == 0:
                idDataK2 = idData
        
        if idx != k1 and idx != k2:
            if len(palavras[idx][1]) > 1 and palavras[idx][1].find("(") >= 0:
                strWord += "class='entity'>"
                strWord += palavras[idx][0]
            elif len(palavras[idx][1]) == 1 and check == True:
                strWord += "class='word'>"
                strWord += palavras[idx][0]
            elif len(palavras[idx][1]) == 1 and check == False:
                strWord += ">"
                strWord += palavras[idx][0]
            elif palavras[idx][1].find("(") < 0:
                strWord += palavras[idx][0]
        
        else:
            if idx == k1:
                strWord += "class='org'>"
                strWord += palavras[idx][0]
            
            else:
                if palavras[idx][1].find("PLC") >= 0:
                    strWord += "class='plc'>"
                    strWord += palavras[idx][0]
                elif palavras[idx][1].find("PER") >= 0:
                    strWord += "class='per'>"
                    strWord += palavras[idx][0]
                else:
                    strWord += "class='org'>"
                    strWord += palavras[idx][0]
            
            check = not check
            
        if palavras[idx][1].find(")") >= 0 or len(palavras[idx][1]) == 1:
            strWord += "</span> "
            
        else:
            strWord += " "
        
        if palavras[idx][1].find(")") >= 0 or len(palavras[idx][1]) == 1:
                strLine = strLine + "<span data-id='" + idData + "'" + strWord  
                strWord = ""
                idData = ""
    
    strLine = idText + "\t" + idSent + "\t" + strLine + "\t" + idDataK1 + "\t" + idDataK2
    
    if idSent == "S1":
        print (strLine)
        print ("")
    with open("saida.tsv", "a") as myfile:
        myfile.write(strLine)
        myfile.write("\n")
    
import collections

open("saida.tsv", 'w').close()
with open("saida.tsv", "a") as myfile:
    myfile.write("id_txt\tid_sent\thtml\tid_ent1\tid_ent2")
    myfile.write("\n")
entidades = collections.OrderedDict()
palavras = collections.OrderedDict()

with open("Summ-it++v3.txt", "r", encoding = "utf8") as file:
    for line in file:
        columns = line.split('\t')
            
        if line.find("#begin") >= 0:
            columns = line.split('/')
            idText = columns[7]
        
        columns = line.split('\t')
        
        if len(columns) >= 6:
            idSent = columns[0]
            palavras[int(columns[1])] = [columns[2], columns[7], columns[8]]
            if columns[7].find("(ORG") >= 0 or columns[7].find("(PLC") >= 0 or columns[7].find("(PER") >= 0:
                entidades[int(columns[1])] = columns[7]

        if line == "\n":
            if len(entidades) > 1:
                for k1, v1 in entidades.items():
                    if v1.find("ORG") >= 0:
                        for k2, v2 in entidades.items():
                            if k1 != k2:
                                if k2 < k1 and v2.find("ORG") >= 0: 
                                    pass
                                else:
                                    montaTexto(idText, idSent, k1, k2, v1, v2, palavras)
            entidades = collections.OrderedDict()
            palavras = collections.OrderedDict()
```
